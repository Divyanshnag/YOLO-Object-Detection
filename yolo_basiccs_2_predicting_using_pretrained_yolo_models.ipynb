```python
from ultralytics import YOLO
model = YOLO("yolo26s.pt") #loading an official
```

```python
# Predicting with the model
results = model("/content/pexels-photo-13872248.jpeg")
```

```python
# Access the results
for i , result in enumerate(results):
  print(f"\nResult {i+1}:")

  xywh = result.boxes.xywh # center-x , center-y , width , height
  print("Bounding Box (xywh)")
  print(xywh)

  xywhn = result.boxes.xywhn # normalised xywh
  print("Bounding Box (xywhn)")
  print(xywhn)

  xyxy = result.boxes.xyxy
  print("Bounding Boxes (xyxy)") # top left x,  top left y , bottom right x , bottom right y
  print(xyxy)

  xyxyn = result.boxes.xyxyn
  print("Normalized Bounding Boxes (xyxyn) : ")  # normalized
  print(xyxyn)

  names = [result.names[cls.item()] for cls in result.boxes.cls.int()]
  print(f"Class Names \n {names}")

  confidence = result.boxes.conf # confidence score of each boxes
  print(f"Confidence Scores : {confidence}")
```

```python
for i , result in enumerate(results):
  print(f"\n\n {i+1}")
  print(f"\n{result.boxes}")
  print(f"\n{result}")
```
